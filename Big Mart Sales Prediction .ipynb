# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df_train.columns
```

```python
df_train=df_train.apply(le.fit_transform)
```

```python
df_train.head()
```

# One Hot Encoding

```python
df_train = pd.get_dummies(df_train, columns=['Item_Fat_Content','Outlet_Location_Type','Outlet_Size','Outlet_Type'])
```

```python
df_train
```

# splitting our data into train and test

```python
X=df_train.drop('Item_Outlet_Sales',axis=1)
```

```python
X
```

```python
Y=df_train['Item_Outlet_Sales']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=101,test_size=0.2)
```

# Standardization

# Model Building

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
Y_pred_lr=lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
print(r2_score(Y_test,Y_pred_lr))
```

```python
print(mean_absolute_error(Y_test,Y_pred_lr))
print(np.sqrt(mean_squared_error(Y_test,Y_pred_lr)))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
```

```python
rf.fit(X_train,Y_train)
```

```python
Y_pred_ref=rf.predict(X_test)
```

```python
print(r2_score(Y_test,Y_pred_ref))
```

```python
print(mean_absolute_error(Y_test,Y_pred_ref))
print(np.sqrt(mean_squared_error(Y_test,Y_pred_ref)))
```
